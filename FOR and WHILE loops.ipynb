```python
values = int(input("How much data do you want to introduce? "))
ages=[] #Here is a table created and waiting for some data  
total=0
for i in range(0,values): #Here depending on the amount you want to add which i assigned in values
    x=int(input("Introduce age: "))#Here any age that we put in, it will send it to the table created above
    ages.append(x)
    total += x #Here every value that you introduce it will change the variable of total to the one you put, and it will sum it with the values that are already sumed in total.
print("This is your sample: ")

for age in ages:
    print(age, end = ' ') #Here it will print every age you inserted 

    sample_mean = total/values
print(f"\nThe Sample Mean is= {sample_mean}")
```

# Excercise Using the FOR,Table and append

We must try to do the same as above but, instead of showing us the mean and age(Numbers), we must change them to names and try to select what we want to see using the tables function.

```python
values = int(input("How many names would you like to introduce? "))
names=[]
for i in range(0,values): 
    x=input("Introduce Names: ")
    names.append(x)
print("These are your names introduced: ")
for name in names:
    print(name)  

print(f"\nPlease choose the number where the name you want to see is positioned,counting from the begining as 0 to {values-1}")
y=int(input())

print(names[y])
```
